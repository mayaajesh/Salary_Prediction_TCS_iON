```python
#Removing employee count,Over18,StandardHours  since these features contains only one value. Dropping EmployeeNumber as all values are unique. 
data.drop(['EmployeeCount','Over18','StandardHours','EmployeeNumber','Attrition','RelationshipSatisfaction',],axis=1,inplace=True)
```

```python
data.dtypes
```

### Encoding

```python
from sklearn.preprocessing import LabelEncoder
label_encoded=LabelEncoder()
for i in data.columns:
    if data[i].dtype == object:
        data[i]=label_encoded.fit_transform(data[i])
```

```python
data.head()
```

### Removing biased features
Obtaining correlation between features

```python
#splitting dataset into features and target. y is the target 
y = data['MonthlyIncome']
x = data.drop('MonthlyIncome', axis = 1)
```

```python
#correlation heatma 
correlate=x.corr()
plt.figure(figsize=(16,16))
ax = sns.heatmap(correlate, annot=True, linewidths=0.1, linecolor='white',cmap="Greens")
```

```python
#Removing highly correlated features
#Features with correlation greater than 0.75 and -0.75 are removing
x.drop(['PercentSalaryHike','YearsWithCurrManager','YearsInCurrentRole', 'JobLevel'],axis=1, inplace=True)
```

### Outlier detection

```python
plt.figure(figsize=(20,7))
x.boxplot()
plt.xticks(rotation=90)
plt.show()
```

```python
#The features with outliers are plotted separately
```

```python
plt.figure(figsize=(4,6))
plt.subplot(2,1,1)
sns.boxplot(y=x['NumCompaniesWorked'])
plt.title('Number of companies worked')
plt.subplot(2,1,2)
plt.title('TrainingTimesLastYear')
sns.boxplot(y=x['TrainingTimesLastYear'])
plt.tight_layout()
plt.show()
sns.boxplot(y=x['StockOptionLevel'])
plt.title('StockOptionLevel')
plt.tight_layout()
plt.show()
sns.boxplot(y=x['YearsAtCompany'])
sns.boxplot(y=x['YearsAtCompany'])
plt.tight_layout()
plt.show()
sns.boxplot(y=x['YearsSinceLastPromotion'])
sns.boxplot(y=x['YearsSinceLastPromotion'])
plt.tight_layout()
plt.show()
```

```python
#Outlier handling
for i in ['NumCompaniesWorked','TrainingTimesLastYear','StockOptionLevel','YearsAtCompany','YearsSinceLastPromotion']:
    q1 = x[i].quantile(0.25) 
    q3 = x[i].quantile(0.75) 
    iqr = q3 - q1 
    upper_limit = q3 + 1.5 * iqr
    lower_limit = q1 - 1.5 * iqr 
#Replacing outliers with upper and lower limit
    x[i][x[i]>upper_limit] = upper_limit
    x[i][x[i]<lower_limit] = lower_limit
```

```python
#Plotting after outlier handling
plt.figure(figsize=(4,6))
plt.subplot(2,1,1)
sns.boxplot(y=x['NumCompaniesWorked'])
plt.title('Number of companies worked')
plt.subplot(2,1,2)
plt.title('TrainingTimesLastYear')
sns.boxplot(y=x['TrainingTimesLastYear'])
plt.tight_layout()
plt.show()
sns.boxplot(y=x['StockOptionLevel'])
plt.title('StockOptionLevel')
plt.tight_layout()
plt.show()
sns.boxplot(y=x['YearsAtCompany'])
sns.boxplot(y=x['YearsAtCompany'])
plt.tight_layout()
plt.show()
sns.boxplot(y=x['YearsSinceLastPromotion'])
sns.boxplot(y=x['YearsSinceLastPromotion'])
plt.tight_layout()
plt.show()
```

## Modeling

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state =42, test_size = 0.2)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

### Logistic Regression

```python
from sklearn.linear_model import LinearRegression
logistic_model = LinearRegression()
logistic_model.fit(x_train,y_train)
y_predict_logistic = logistic_model.predict(x_test)
print("R squared value of Logistic Regression is : ",r2_score(y_test,y_predict_logistic))

```

### KNN

```python
from sklearn.neighbors import KNeighborsRegressor
r_value = []
#creating neighbors as 3 to 15
neighbors = np.arange(3,15)
for k in neighbors:
    # taking instances for KNN into variable name classifier using the method minkowski
    knn_r = KNeighborsRegressor(n_neighbors=k, metric="minkowski")
    knn_r.fit(x_train,y_train)
    y_pred_knn = knn_r.predict(x_test)
    r2 = r2_score(y_test,y_pred_knn)
    r_value.append(r2)
```

```python
r_value
```

```python
#since this model gives negative r2 value, not considering this model
```

### SVM

```python
from sklearn.svm import SVR
svm_regressor=SVR(kernel='rbf')
svm_regressor.fit(x_train,y_train)
y_pred_svm=svm_regressor.predict(x_test)
print("R squared value of SVM is : ",r2_score(y_test,y_pred_svm))
```

```python
#This model also gives negative r2 value, so not considering this model
```

### Random Forest 

```python
from sklearn.ensemble import RandomForestRegressor
random_forest=RandomForestRegressor()
random_forest.fit(x_train,y_train)
y_pred_random_forest=random_forest.predict(x_test)
print("R squared value of Random Forest model is : ",r2_score(y_test,y_pred_random_forest))
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
decision_tree=DecisionTreeRegressor()
decision_tree.fit(x_train,y_train)
y_pred_decision_tree=decision_tree.predict(x_test)
print("R squared value of Decision Tree is : ",r2_score(y_test,y_pred_decision_tree))
```

### Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingRegressor
grad_boost=GradientBoostingRegressor()
grad_boost.fit(x_train, y_train)
y_pred_grad_boost=grad_boost.predict(x_test)
print("R squared value of gradient boosting is : ",r2_score(y_test,y_pred_grad_boost))
```

### XGBoost

```python
from xgboost import XGBRegressor
xg_boost=XGBRegressor()
xg_boost.fit(x_train,y_train)
y_pred_xg_boost=xg_boost.predict(x_test)
print("R squared value  of XG Boost is : ",r2_score(y_test,y_pred_xg_boost))
```

```python
#Gradient Boost gives good r2 score. Therefore it can be used for modelling
```
